```python
import pandas as pd #Library to handle with dataframes
import matplotlib.pyplot as plt # Library to plot graphics
import numpy as np # To handle with matrices
import seaborn as sns # to build modern graphics
from scipy.stats import kurtosis, skew # it's to explore some statistics of numerical values
```

```python
file_to_output = "wine_data.txt"
```

```python
df = pd.read_csv('wine_data.csv', index_col=0)
df.head()
```

```python
# Print the results and export the data to our text file
for item in df["description"].values[:5000]:
    with open(file_to_output, "a") as txt_file:

        description = (
#         f"\n\nDescription\n"
           f"-------------------------\n"
           f"{item}\n"
           f"-------------------------\n")
        print(description, end="")

       # Save to a .txt file
        txt_file.write(description)
```

```python
import os
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS

text = open(path.join('wine_data.txt')).read()

stopwords = set(STOPWORDS)
stopwords.update(["drink", "now", "wine", "flavor", "flavors", "description"])

wordcloud = WordCloud().generate(text)

import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis("off")

wordcloud = WordCloud(background_color = 'white', stopwords=stopwords, max_font_size = 40, max_words =150).generate(text)
plt.figure(figsize = [20,10])
plt.imshow(wordcloud, interpolation = "bilinear")
plt.axis("off")
plt.savefig('wordcloud.png')
plt.show
```
