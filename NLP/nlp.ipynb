```python
import pandas as pd
from rake_nltk import Rake
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python

df = pd.read_csv('winemag-data-130k-v2.csv')
df = df[["country", "description", 'variety', 'winery']]
```

```python
df = df.dropna(how='any')
```

```python
df['Key_words'] = ""
for index, row in df.iterrows():
    description = row['description']
    r = Rake()

    r.extract_keywords_from_text(description)

    key_words_dict_scores = r.get_word_degrees()
    
    row['Key_words'] = list(key_words_dict_scores.keys())

df.drop(columns = ['description'], inplace = True)
```

```python
df2 = df[:5000]
df3 = df2[["Key_words", "variety"]]
df3.head()
```

```python

df3['Key_words']=[" ".join(review) for review in df3['Key_words'].values]
```

```python
print(type(df3["Key_words"][0]))
print(type(df3["Key_words"][0][0]))
```

```python
count = CountVectorizer()
count_matrix = count.fit_transform(df3["Key_words"])
cosine_sim = cosine_similarity(count_matrix, count_matrix)
```

```python
df3 = df.set_index('variety')
df3.head()
```

```python
title = "Riesling"
```

```python
indices = pd.Series(df3.index)


def recommendations(title, cosine_sim = cosine_sim):
    

    recommended_wines = []
    

    idx = indices[indices == title].index[0]


    score_series = pd.Series(cosine_sim[idx]).sort_values(ascending = False)


    top_5_indexes = list(score_series.iloc[1:6].index)
    
  
    for i in top_5_indexes:
        recommended_wines.append(list(df3.index)[i])
        
    print(recommended_wines)
```

```python
recommendations(title)
```
